#### This scrapes the data from IMDB
* Loops through the last 18 years of Oscar pages
* Gets the list of movies
* What oscar categories were they nominated for
* What golden globes were they nominated for and what ones did they win
* gross
* average review

```python
from bs4 import BeautifulSoup
import requests
```

```python
url_base = ['https://www.imdb.com/title/', '/awards?ref_=tt_ql_op_1']
list_movies = ['tt0133093', 'tt3783958']
list_awards = ['Golden Globe', 'Oscar']
```

```python
for movie in list_movies:
    print(movie)

    res_movie = []
    # gets the movie awards page
    url_year = url_base[0] + movie + url_base[1]
    res = requests.get(url_year)
    soup = BeautifulSoup(res.text)
    
    # finds the tables and for each award type we loop through and get info
    tables = soup.find_all('table', class_='awards')
    for award in list_awards:
        winner = False
        for table in tables:
            if table.find('span').text == award:
                for td in table.find_all('td'):
                    if td.text.replace(award, '').strip() == 'Winner':
                        winner = True                    
                    elif td.text.replace(award, '').strip() == 'Nominee':
                        winner = False
                    elif td['class'][0] == 'award_description':
                        text = td.text.split('\n')
                        res_movie.append([movie, award, text[1].strip(), winner])
```
